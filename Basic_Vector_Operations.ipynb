Addition of Vectors

```python
v_1 = np.asarray([0,0,2,4])
v_2 = np.asarray([0,0,3,-2])
sum = v_1 + v_2
print(sum)
```

Subtraction of Vectors

```python
v_1 = np.asarray([0,0,2,4])
v_2 = np.asarray([0,0,3,-2])
sum = v_1 - v_2
print(sum)
```

```python
plt.quiver(0,0,2,4,angles='xy',scale_units='xy',scale=1,color='b')
plt.quiver(0,0,3,-2,angles='xy',scale_units='xy',scale=1,color='g')
plt.quiver(0,0,-1,6,angles='xy',scale_units='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Multiply a vector by a scalar

```python
v_1 = np.asarray([0,0,2,4])
v_2 = 2*v_1
print(v_2)
```

Dot Product of 2 Vectors

```python
a = np.asarray([2,3])
b = np.asarray([4,4])
dot = np.dot(a,b)
print(dot)
```

Cross-Product

```python
c = np.array([5,10,20])
d = np.array([10,20,30])
cross = np.cross(c,d)
print(cross)
#
```

Projection of A vector on V

```python
#projection = (a.v/||v||^2)v
```

```python
a = np.array([2,5])
v = np.array([8,-6])
magnitude_v = np.linalg.norm(v)
print(magnitude_v)
#projection
projection = (np.dot(a,v)/magnitude_v**2)*v
print(projection)
```

```python
matrix_ones = np.ones((3,3),dtype=int)
print(matrix_ones)
```
